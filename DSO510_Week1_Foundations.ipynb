## Lists
Demonstration of list creation, indexing, slicing, and mutation using variables
from the vibe coding content analysis project.

```python
# List of platforms in the vibe coding study
platforms = ["YouTube", "Reddit"]

# List of sentiment categories used in coding
sentiment_categories = ["positive", "negative", "neutral", "mixed"]

# List of unit IDs analyzed
unit_ids = [1, 2, 3, 4, 5]

print(platforms)
print(sentiment_categories)
print(unit_ids)
```

```python
print(sentiment_categories[0])
print(sentiment_categories[1])
print(sentiment_categories[3])
```

```python
# How many items are in a list
print(len(sentiment_categories))

# Adding an item to a list
sentiment_categories.append("unclear")
print(sentiment_categories)

# Accessing the last item without knowing the index
print(sentiment_categories[-1])
```

```python
# Slicing a list — grabbing a subset
print(sentiment_categories[0:3])
print(sentiment_categories[1:4])
print(sentiment_categories[:2])
print(sentiment_categories[2:])
```

## Loops and Conditionals
Demonstration of for loops and if/elif/else conditionals applied to
sentiment score classification. Logic mirrors real coding decisions
made during the CMGT 514 vibe coding content analysis.

```python
for category in sentiment_categories:
    print(category)
```

```python
for category in sentiment_categories:
    print(f"Sentiment category: {category}")
```

```python
for category in sentiment_categories:
    if category == "negative":
        print(f"WARNING: {category} sentiment detected")
    else:
        print(f"{category} sentiment is fine")
```

```python
for category in sentiment_categories:
    if category == "positive":
        print(f"{category}: favorable toward vibe coding")
    elif category == "negative":
        print(f"{category}: critical of vibe coding")
    elif category == "neutral":
        print(f"{category}: neither favorable nor critical")
    else:
        print(f"{category}: requires further review")
```

```python
# Sentiment scores for 10 coded units
sentiment_scores = [0.82, 0.14, 0.46, 0.55, 0.91, 0.23, 0.67, 0.38, 0.74, 0.11]

for score in sentiment_scores:
    if score >= 0.70:
        print(f"{score}: High positive sentiment")
    elif score >= 0.50:
        print(f"{score}: Leaning positive sentiment")
    elif score >= 0.40:
        print(f"{score}: Moderate sentiment")
    else:
        print(f"{score}: Low/negative sentiment")
```

## Dictionaries
Demonstration of dictionary structure — key/value pairs used to store
connected information. Here, platform names are connected to their
unit counts from the vibe coding content analysis.

```python
# Units analyzed per platform
platform_counts = {
    "YouTube": 82,
    "Reddit": 68
}

print(platform_counts)
```

```python
# Accessing values by key
print(platform_counts["YouTube"])
print(platform_counts["Reddit"])

# This is different from a list — you use the key, not an index number
print(f"YouTube units: {platform_counts['YouTube']}")
print(f"Reddit units: {platform_counts['Reddit']}")
```

```python
# Same information stored two different ways

# As a list — just the platform names in order
platforms_list = ["YouTube", "Reddit"]

# As a dictionary — platform names connected to their unit counts
platforms_dict = {
    "YouTube": 82,
    "Reddit": 68
}

# Accessing from a list — you use position
print(platforms_list[0])

# Accessing from a dictionary — you use the key name
print(platforms_dict["YouTube"])
```
